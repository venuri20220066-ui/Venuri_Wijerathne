**Data Pre-processing**

```python
df = df.drop(columns=["Timestamp"])
```

```python
df = df.dropna(how='all')
```

```python
df.columns = df.columns.str.strip()
# Remove leading and trailing spaces from column names

df.columns = df.columns.str.replace("\n", "", regex=True)
# Remove newline characters (\n) from column names to make them clean and consistent
```

```python
df.head()
```

```python
df = df.rename(columns={
    'What is your gender?': 'gender',
    'What is your marital status?': 'marital_status',
    'Have you ever failed an academic course/module at university?': 'academic_failure',
    'How satisfied are you with your learning environment at the Faculty?': 'faculty_environment',
    'Are you satisfied with your home environment for studying?': 'home_environment',
    'What is your Age?': 'age_group',
    'What is your Academic Year?': 'academic_year',
    'What is your degree programme.': 'degree_program',
    'Competition with my peers for grades is quite intense': 'peer_competition',
    'I feel my lecturers are very strict and often point out mistakes in my work': 'lecturer_strictness',
    'My lecturers expect more from me than I can realistically do.': 'lecturer_expectations',
    'The unrealistic expectations of my parents stress me out.': 'parental_expectations',
    'The time allocated for lectures, laboratory sessions, and independent study is enough for me to manage my academic workload effectively.': 'time_management',
    'The size of the curriculum (workload) is excessive.': 'curriculum_load',
    'I believe that the amount of work assignments is too much.': 'assignment_load',
    'I am unable to catch up if I fall behind in my work.': 'difficulty_catching_up',
    'I have enough time to relax after completing my academic work.': 'relaxation_time',
    'The examination questions are usually difficult.': 'exam_difficulty',
    'The allocated examination duration is too short for me to complete my answers properly.': 'exam_time_pressure',
    'Examination periods are very stressful to me.': 'exam_stress',
    'I am confident that I will be a successful student.': 'academic_confidence',
    'I am confident that I will be successful in my future career.': 'career_confidence',
    'I find it easy to make academic decisions.': 'decision_making',
    'I fear failing courses/modules this year.': 'module_failure',
    'I think that my worry about examinations is weakness of character.': 'exam_concern',
    'Even if I pass my exams, I am worried about getting a job': 'employment_concern'
})
```

```python
pd.DataFrame(df.columns, columns=["Column Names"])
```

```python
reverse_items = [
    "peer_competition",
    "lecturer_strictness",
    "lecturer_expectations",
    "parental_expectations",
    "curriculum_load",
    "assignment_load",
    "difficulty_catching_up",
    "exam_difficulty",
    "exam_time_pressure",
    "exam_stress",
    "module_failure",
    "exam_concern",
    "employment_concern"
]

# Apply reverse coding (1â5 Likert scale)
for col in reverse_items:
    df[col] = 6 - df[col]
```

```python
df.head()
```

```python
# Columns to clean (remove weird symbols like Â)
cols_to_clean = [
    "marital_status",
    "academic_failure",
    "home_environment",
    "academic_year"
]
for col in cols_to_clean:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r"[^\w\s]", "", regex=True)  # remove weird symbols
        .str.strip()
    )
df["age_group"] = (
    df["age_group"]
    .astype(str)
    .str.replace(r"[^\d]", "", regex=True)   # keep only numbers
    .str.replace(r"(\d{2})(\d{2})", r"\1-\2", regex=True)  # insert hyphen
)
```

```python
df.head()
```

```python
df["degree_program"] = df["degree_program"].str.strip()
df["degree_program"] = df["degree_program"].str.replace(r"\s+", " ", regex=True)
```

```python
# Fix inconsistent degree program names
df["degree_program"] = df["degree_program"].replace({
    "B.B.A. (Hons) in Hospitality and Leisure Mnagaement": "B.B.A. (Hons) in Hospitality and Leisure Management",
    "B.B.A. (Hons) in Hospitality and Leisure Mnaga...": "B.B.A. (Hons) in Hospitality and Leisure Management"
}, regex=True)
```

```python
# Standardize using keyword matching
df["degree_program"] = df["degree_program"].str.replace(
    r".*Hospitality and Leisure.*",
    "B.B.A. (Hons) in Hospitality and Leisure Management",
    regex=True
)
```
