```python
import pandas as pd
df_encoded = pd.read_csv('/content/drive/MyDrive/FYP_IIT/Cleaned_Prediction_Dataset.csv', encoding='latin1')
df_encoded.head()
```

```python
print(df_encoded.columns.tolist())
```

```python
y = df_encoded[['F1', 'F2', 'F3', 'F4']]
```

```python
X = df_encoded[
    ['gender_Female', 'gender_Male',
     'academic_failure_Never', 'academic_failure_Once or twice',
     'academic_failure_Three or more times',
     'faculty_environment_Neither satisfied nor dissatisfied',
     'faculty_environment_Not satisfied', 'faculty_environment_Satisfied',
     'faculty_environment_Very satisfied',
     'home_environment_No', 'home_environment_Yes',
     'age_group_20-21', 'age_group_22-23', 'age_group_24 and above',
     'academic_year_Year 1', 'academic_year_Year 2',
     'academic_year_Year 3', 'academic_year_Year 4',
     'degree_program_B.B.A. (Hons) in Accounting',
     'degree_program_B.B.A. (Hons) in Business Economics',
     'degree_program_B.B.A. (Hons) in Finance',
     'degree_program_B.B.A. (Hons) in International Business',
     'degree_program_B.B.A. (Hons) in Marketing',
     'degree_program_General', 'degree_program_Other']
]
```

```python
from sklearn.model_selection import train_test_split
# Import function to split dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X, y,                # X = input features, y = target variables (e.g., F1âF4)
    test_size=0.2,       # 20% of data used for testing, 80% for training
    random_state=42      # Ensures the same split every time (reproducibility)
)
```

```python
from xgboost import XGBRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import r2_score
```

```python
from xgboost import XGBRegressor  # Import XGBoost regression model
from sklearn.multioutput import MultiOutputRegressor  # Wrapper for multi-target prediction

xgb = XGBRegressor(
    n_estimators=100,     # Number of boosting trees (more trees = potentially better performance but slower)
    learning_rate=0.1,    # Step size shrinkage to prevent overfitting
    max_depth=4,          # Maximum depth of each tree (controls model complexity)
    random_state=42       # Ensures reproducibility of results
)

model = MultiOutputRegressor(xgb)
# Wrap the XGBRegressor to handle multiple output variables (predicting F1, F2, F3, F4 together)
```

```python
model.fit(X_train, y_train)
# Train the MultiOutput XGBoost model using the training data
# X_train = input features used to learn patterns
# y_train = target variables (e.g., F1âF4 stress dimensions)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
print("RÂ² Score:", r2)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
# RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# MAE
mae = mean_absolute_error(y_test, y_pred)

print("RMSE:", rmse)
print("MAE:", mae)
```

```python
# RMSE per target
rmse_each = np.sqrt(mean_squared_error(y_test, y_pred, multioutput='raw_values'))

# MAE per target
mae_each = mean_absolute_error(y_test, y_pred, multioutput='raw_values')

print("RMSE (F1, F2, F3, F4):", rmse_each)
print("MAE  (F1, F2, F3, F4):", mae_each)
```

```python
from sklearn.metrics import r2_score

print("Overall RÂ²:", r2_score(y_test, y_pred))
print("Overall RMSE:", rmse)
print("Overall MAE:", mae)

print("\nRÂ² (each):", r2_score(y_test, y_pred, multioutput='raw_values'))
print("RMSE (each):", rmse_each)
print("MAE (each):", mae_each)
```
